```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cust_info = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

## step1: Data Cleaning

```python
cust_info.head()
```

```python
cust_info.shape, cust_info.info
```

```python
cust_info.isnull().sum()
```

Ther are no null values in the dataset

```python
cust_info.dtypes
```

```python
cust_info['TotalCharges'] = pd.to_numeric(cust_info['TotalCharges'], errors='coerce')
```

```python
cust_info.dtypes
```

```python
cust_info.head()
```

```python
cust_info['TotalCharges'].isnull().sum()
```

```python
cust_info["TotalCharges"]=cust_info["TotalCharges"].fillna(cust_info["TotalCharges"].median())
```

```python
cust_info.dtypes
```

```python
cust_info['TotalCharges'].isnull().sum()
```

```python
cust_info.head()
```

#### dropping customer id since its not a feature

```python
cust_info=cust_info.drop("customerID", axis=1)
```

#### converting churn to numeric values

```python
cust_info["Churn"]=cust_info["Churn"].map({"Yes":1,"No":0})
```

```python
cust_info.head()
```

```python
cust_info.duplicated().sum()
```

```python
for col in cust_info.select_dtypes(include='object').columns:
    print(col, '→', cust_info[col].unique())
```

## 2. Calculations

```python
cust_info.head()
```

```python
cust_info["Churn"].value_counts(normalize=True)*100
```

```python
## churn rate by contract type
cust_info.groupby("Contract")["Churn"].mean()*100
```

```python
## churn by internet service
cust_info.groupby("InternetService")["Churn"].mean()*100
```

```python
## average tenure of churned and non churned customer
cust_info.groupby("Churn")["tenure"].mean()
```

```python
## average monthly charges paid by churned and non churned customers
cust_info.groupby("Churn")["MonthlyCharges"].mean()
```

```python
## relation of other features with churn (near one means churn value goes up if the feture value goes up near -1 means churn value goe sdown if the feature vale goes up)
cust_info.select_dtypes(include='number').corr()['Churn'].sort_values()
```

## Graphs

```python
%matplotlib inline
cust_info["Churn"].value_counts().plot(kind="bar")
plt.title("Churn distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show();
```

```python
cf=pd.crosstab(cust_info["Churn"],cust_info["Contract"])
cf.plot(kind= "bar")
plt.xlabel("Churn")
plt.ylabel("Churn count")
plt.title("Churning by contract")
plt.show()
```

```python
cf=pd.crosstab(cust_info["Churn"],cust_info["InternetService"])
cf.plot(kind= "bar")
plt.xlabel("Churn")
plt.ylabel("Churn count")
plt.title("Churning by InternetService")
plt.show()
```

```python
plt.hist(cust_info[cust_info['Churn'] == 0]['tenure'], bins=30, alpha=0.6, label='No churn', color="blue")
plt.hist(cust_info[cust_info['Churn'] == 1]['tenure'], bins=30, alpha=0.6, label='Churn', color="red")

plt.title('Tenure distribution by churn')
plt.xlabel('Tenure (months)')
plt.ylabel('Number of customers')
plt.legend()
plt.show()
```

```python
plt.hist(cust_info[cust_info['Churn'] == 0]['MonthlyCharges'], bins=30, alpha=0.6, label='No churn', color="blue")
plt.hist(cust_info[cust_info['Churn'] == 1]['MonthlyCharges'], bins=30, alpha=0.6, label='Churn', color="red")

plt.title('Monthly charges distribution by churn')
plt.xlabel('Monthly charges')
plt.ylabel('Number of customers')
plt.legend()
plt.show()
```

```python
import numpy as np

corr_matrix = cust_info.select_dtypes(include='number').corr()

fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)

ax.set_xticks(range(len(corr_matrix.columns)))
ax.set_xticklabels(corr_matrix.columns, rotation=90)
ax.set_yticks(range(len(corr_matrix.columns)))
ax.set_yticklabels(corr_matrix.columns)

fig.colorbar(im)
plt.title('Feature correlation heatmap')
plt.tight_layout()
plt.show()
```
